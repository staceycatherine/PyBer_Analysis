```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
# can also be written like this
# fig = plt.figure()
# ax = fig.add_subplot()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(). This is using the object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis) #same result
```

```python
# IMPORTANT- to change figure-level attributes(labels, title, legend) OR save as an IMAGE fig = plt.figure()
# Create the plot-use plt.show()
plt.plot(x_axis, y_axis)
plt.show() # plt.show should only be used once per session because it can cause the graph to display improperly
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title
plt.title("Pyber Fare by Month")

# Add the legend.
plt.legend()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```
